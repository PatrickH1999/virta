```python
import numpy as np
import matplotlib.pyplot as plt

def load_field2d(filename):
    with open(filename, "rb") as f:
        ni = np.fromfile(f, dtype=np.uint64, count=1)[0]
        nj = np.fromfile(f, dtype=np.uint64, count=1)[0]

        data = np.fromfile(f, dtype=np.float64)

    arr = data.reshape((nj, ni))
    return arr

arr = load_field2d("output.npy")
print(arr)

plt.imshow(arr, aspect="auto")
plt.colorbar()
plt.title("Field2D")
plt.show()
```
